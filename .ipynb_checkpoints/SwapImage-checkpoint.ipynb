## Swap the Face Of Lucifer
### In this Task we have two images of Lucifer one in Devil and another one normal we will swap the image faces.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
DevilPic = cv2.imread('./images/Lucifer.png')
#cv2.imshow('Frame1',DevilPic)
#cv2.waitKey()
#cv2.destroyAllWindows()
plt.imshow(DevilPic)
shap=DevilPic.shape
shap
```

```python
cDevilPic=DevilPic[50:450,350:600]
#cv2.imshow('CutFace',cDevilPic)
#cv2.waitKey()
#cv2.destroyAllWindows()
plt.imshow(cDevilPic)
cDevilPic.shape
```

```python
NormalPic= cv2.imread('./images/luciferNormal.jpg')
plt.imshow(NormalPic)
NormalPic.shape
```

```python
#Lets Resize them in the Same pixel
NormalPic = cv2.resize(NormalPic,(722,952))
plt.imshow(NormalPic)
NormalPic.shape
```

```python
cNormalPic= NormalPic[10:440,250:450]
#cv2.imshow('Frame2',NormalPic)
#cv2.waitKey()
#cv2.destroyAllWindows()
plt.imshow(cNormalPic)


cNormalPic = cv2.resize(cNormalPic,(250,400))
plt.imshow(cNormalPic)
cNormalPic.shape
```

```python
DevilPic[50:450,350:600]=cNormalPic
#cv2.imshow('CutFace',NormalPic)
#cv2.waitKey()
#cv2.destroyAllWindows()
#NormalPic=cNormalPic
plt.imshow(DevilPic)
```
